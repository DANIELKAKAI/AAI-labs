```python
import matplotlib.pyplot as plt
countries = decade_gdp_df['Country']
years = decade_gdp_df.columns[1:]

for index, row in decade_gdp_df.iterrows():
    country_name = row['Country']
    gdp_values = row[years]

    plt.plot(years, gdp_values, label=country_name)
    
plt.xlabel('Year')
plt.ylabel('Gross domestic product per capita (US Dollars)')
plt.title('Gross domestic product per capita Over the last 10 Years')
plt.legend()
plt.grid(True)
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))

plt.show()
```
