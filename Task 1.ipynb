```python
import pandas as pd
```

```python
df = pd.read_excel('WEOOct2020all.xls', engine='xlrd')
```

```python
gdp_df = df[(df['Subject Descriptor'].str.contains('Gross domestic product per capita', case=False, na=False)) & (df['Units'].str.match('U.S. dollars', case=False, na=False)) & (df['Scale'].str.match('Units', case=False, na=False))]
```

```python
%store gdp_df
```

```python
from datetime import datetime
current_year = datetime.now().year
columns_to_keep = ['Country'] + [year for year in range(current_year - 10, current_year + 1)]
decade_gdp_df = gdp_df[columns_to_keep].copy()
%store decade_gdp_df
```

```python
decade_gdp_df['Difference'] = decade_gdp_df[2023] - decade_gdp_df[2013]
top_10_gdp = decade_gdp_df.sort_values(by='Difference', ascending=False).head(10)
```
